# Training the model using Linear Regression Algorithm

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
features = ['Age', 'Gender', 'Hours_Studied', 'IQ']
```

```python
le = LabelEncoder()
data['Has_Part_Time_Job'] = le.fit_transform(data['Has_Part_Time_Job'])
data['Study_Hours_Group'] = le.fit_transform(data['Study_Hours_Group'])
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
X = data[features]
Y = data[['Physics_Marks','Math_Marks','Chemistry_Marks']]
```

```python
# Training: 80%, Testing: 20%
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, Y_train)
prediction = model.predict(X_test)
```

```python
age = float(input("Enter your age: "))
gender = input("Enter your gender: ")
study_hours = float(input("Enter hours studied: "))
iq = float(input("Enter your IQ: "))
gender_encode = le.fit_transform([gender])[0]
user_input = np.array([age, gender_encode, study_hours, iq]).reshape(1,-1)
mark_pred = model.predict(user_input)
```

```python
print("Predicted Physics Marks:", mark_pred[0, 0])
print("Predicted Math Marks:", mark_pred[0, 1])
print("Predicted Chemistry Marks:", mark_pred[0, 2])
```
